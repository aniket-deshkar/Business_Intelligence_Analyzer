###  Merging Datasets

```python
#Merging customer_dataset with orders_dataset to get a comprehensive view of customer orders
customer_orders = pd.merge(
    customers_dataset, 
    orders_dataset, 
    on='customer_id', 
    how='left')
customer_orders.info()

#Merging order_items with order_items dataset to get a comprehensive view of the products ordered
customer_order_items = pd.merge(
    customer_orders, 
    order_items_dataset, 
    on='order_id', 
    how='left')
customer_order_items.info()

#Merging customer_order_items with products dataset to get a comprehensive view of the products ordered
customer_order_items_products = pd.merge(
    customer_order_items, 
    products_dataset, 
    on='product_id', 
    how='left')
customer_order_items_products.info()

#Merging customer_order_items_products with sellers dataset to get a comprehensive view of the products ordered and their sellers
customer_order_items_products_sellers = pd.merge(
    customer_order_items_products, 
    sellers_dataset, 
    on='seller_id', 
    how='left')
customer_order_items_products_sellers.info()
```

```python
#Merging customer_order_items_products_sellers with order_payments dataset to get a comprehensive view of the products ordered, their sellers and the payments made
customer_order_items_products_sellers_payments = pd.merge(
    customer_order_items_products_sellers, 
    order_payments_dataset, 
    on='order_id', 
    how='left')
customer_order_items_products_sellers_payments.info()
```

```python

#Merging customer_order_items_products_sellers_payments with order_reviews dataset to get a comprehensive view of the products ordered, their sellers, the payments made and the reviews given
customer_order_items_products_sellers_payments_reviews = pd.merge(  
    customer_order_items_products_sellers, 
    order_reviews_dataset, 
    on='order_id', 
    how='left')
customer_order_items_products_sellers_payments_reviews.info()
```

```python
#Category translation
customer_order_items_products_sellers_category_translation = pd.merge(
    customer_order_items_products_sellers, 
    product_category_name_translation, 
    on='product_category_name', 
    how='left')
customer_order_items_products_sellers_category_translation.info()
```

```python
# Dataset for analysis/modeling
# Use the merged dataframe that includes reviews so review_score is available
linear_data = customer_order_items_products_sellers_payments_reviews.copy()
data = linear_data.copy()
data.info()
```

```python
# Drop columns not needed for your final model after feature engineering if required
# Example:
# cols_to_drop = ['customer_unique_id', 'order_id', 'customer_id']
# data = data.drop(columns=[c for c in cols_to_drop if c in data.columns])
cols_to_keep = [
    'customer_id',
    'customer_city',
    'customer_state',
    'order_status',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'order_item_id',
    'product_id',
    'seller_id',
    'shipping_limit_date',
    'price',
    'freight_value',
    'product_category_name',
    'seller_zip_code_prefix',
    'seller_city',
    'seller_state',
    'payment_sequential',
    'payment_type',
    'payment_installments',
    'payment_value',
    'order_purchase_timestamp',
    'review_score',
]

linear_data = data.drop(
    columns=[col for col in data.columns if col not in cols_to_keep]
)
print(linear_data.info())
print(linear_data.columns)
```

### Feature Engineering

```python
# Fixing date columns
date_columns = [
    'order_purchase_timestamp',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in date_columns:
    if col in linear_data.columns:
        linear_data[col] = pd.to_datetime(linear_data[col], errors='coerce')
```

```python
#Calculating actual delivery time in days
linear_data['actual_delivery_time'] = (linear_data['order_delivered_customer_date'] - linear_data['order_purchase_timestamp']).dt.days
```

```python
# Calculating late deliveries
# Ensure estimated delivery and late flag are computed using the cleaned dataframe
linear_data['estimated_delivery_time'] = (linear_data['order_estimated_delivery_date'] - linear_data['order_purchase_timestamp']).dt.days
linear_data['late_delivery'] = linear_data['actual_delivery_time'] > linear_data['estimated_delivery_time']
```

```python
#Frieght Ratio 
linear_data['freight_ratio'] = linear_data['freight_value'] / linear_data['price']
print(linear_data['freight_ratio'].describe())
```

```python
#Delivery time distribution
linear_data['actual_delivery_time'].hist(bins=30)
plt.title('Distribution of Actual Delivery Time')
# plt.figure(figsize=(10,6))
plt.xlabel('Actual Delivery Time (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Review Score vs Delay 
linear_data['review_score'] = linear_data['review_score'].fillna(0).astype(int)  # Filling NaN review scores with 0 for analysis
sns.boxplot(x='late_delivery', y='review_score', data=linear_data)
plt.title('Review Score vs Late Delivery')
plt.xlabel('Late Delivery')
plt.ylabel('Review Score')
plt.show()
```

```python
#Adding sales column 
linear_data['sales'] = linear_data['price'] + linear_data['freight_value']
```

```python
# Improved: Aggregate sales with richer features
city_state_sales = (
    linear_data
    .groupby(['customer_state', 'customer_city'], as_index=False)
    .agg(
        total_sales=('sales', 'sum'),
        num_orders=('order_status', 'count'),
        avg_order_value=('sales', 'mean'),
        avg_price=('price', 'mean'),
        avg_freight=('freight_value', 'mean'),
        median_review=('review_score', 'median'),
    )
)

# Fill any NaN reviews with the median
city_state_sales['median_review'] = city_state_sales['median_review'].fillna(
    city_state_sales['median_review'].median()
)

y = city_state_sales['total_sales']
print('Aggregated shape:', city_state_sales.shape)
print(city_state_sales.head())
```

```python
# Improved: Use aggregate numeric features + one-hot encode ONLY state (27 values)
# NOT one-hot encoding customer_city (~4000 values was causing overfitting)

X = city_state_sales[[
    'customer_state',
    'num_orders',
    'avg_order_value',
    'avg_price',
    'avg_freight',
    'median_review',
]].copy()

# One-hot encode only customer_state (27 unique values — manageable)
X_encoded = pd.get_dummies(
    X,
    columns=['customer_state'],
    drop_first=True

# Fill any remaining NaN values\n
)
X_encoded = X_encoded.fillna(0)


# Fill any remaining NaN values
X_encoded = X_encoded.fillna(0)

print('Feature matrix shape:', X_encoded.shape)
print('NaN count:', X_encoded.isna().sum().sum())
print(X_encoded.head())
```

### Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, root_mean_squared_error

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate
r2 = r2_score(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)

print('=' * 50)
print('LINEAR REGRESSION RESULTS')
print('=' * 50)
print(f'R²:   {r2:.4f}')
print(f'RMSE: {rmse:.2f}')
print()
print('Previous scores:')
print('  R²:   -0.1484')
print('  RMSE: 12071.61')
print()
print(f'R² improvement:   {r2 - (-0.1484):.4f}')
print(f'RMSE improvement: {12071.61 - rmse:.2f}')
```

```python
# ── Visualization: Linear Regression Diagnostics ──
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-whitegrid')

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1. Actual vs Predicted scatter
axes[0].scatter(y_test, y_pred, alpha=0.6, s=30, c='#2196F3', edgecolors='white', linewidth=0.3)
max_val = max(y_test.max(), max(y_pred))
axes[0].plot([0, max_val], [0, max_val], 'r--', lw=2, label='Perfect Prediction')
axes[0].set_xlabel('Actual Sales (R$)', fontsize=12)
axes[0].set_ylabel('Predicted Sales (R$)', fontsize=12)
axes[0].set_title(f'Actual vs Predicted Sales\nR² = {r2:.4f}  |  RMSE = {rmse:.0f}', fontsize=13, fontweight='bold')
axes[0].legend(fontsize=11)

# 2. Residuals histogram
residuals = y_test.values - y_pred
axes[1].hist(residuals, bins=40, color='#FF7043', edgecolor='white', alpha=0.85)
axes[1].axvline(x=0, color='black', linestyle='--', lw=1.5)
axes[1].set_xlabel('Residual (Actual − Predicted)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_title('Residuals Distribution', fontsize=13, fontweight='bold')

plt.tight_layout(pad=3)
plt.show()
```

```python
# ── State-wise Prediction Analysis ──

# Map test set back to state names
test_indices = X_test.index
state_test = city_state_sales.loc[test_indices, 'customer_state'].values

state_results = pd.DataFrame({
    'State': state_test,
    'Actual': y_test.values,
    'Predicted': y_pred,
})
state_results['Error'] = (state_results['Actual'] - state_results['Predicted']).abs()

# Aggregate by state
state_summary = state_results.groupby('State').agg(
    Cities=('Actual', 'count'),
    Actual_Total=('Actual', 'sum'),
    Predicted_Total=('Predicted', 'sum'),
    Mean_Error=('Error', 'mean'),
).reset_index()

state_summary = state_summary.sort_values('Actual_Total', ascending=False)

print('State-wise Prediction Summary (sorted by sales):')
print(state_summary.to_string(index=False, float_format='{:,.0f}'.format))

# Plot only top 10 states for clarity
top10 = state_summary.head(10).sort_values('Actual_Total', ascending=True)

fig, ax = plt.subplots(figsize=(12, 6))

y_pos = np.arange(len(top10))
bar_height = 0.35

bars1 = ax.barh(y_pos - bar_height/2, top10['Actual_Total'], bar_height,
                label='Actual Sales', color='#2196F3', edgecolor='white')
bars2 = ax.barh(y_pos + bar_height/2, top10['Predicted_Total'], bar_height,
                label='Predicted Sales', color='#FF7043', edgecolor='white')

ax.set_yticks(y_pos)
ax.set_yticklabels(top10['State'], fontsize=12, fontweight='bold')
ax.set_xlabel('Total Sales (R$)', fontsize=12)
ax.set_title('Top 10 States: Actual vs Predicted Sales', fontsize=14, fontweight='bold')
ax.legend(fontsize=11, loc='lower right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
